```python
import re
import json
import requests
import pandas as pd
import datetime as dt
import time

from Shopee_crawl import ShopeeCrawler
```

```python
crawler = ShopeeCrawler()
products = pd.read_csv('D:\\NLP\\Shopee\\product_full_duc.csv')[["product_url", "name", "brand", "industry", "price", "shipping_info", 'product_type_lv1',
       'product_type_lv2', 'origin', 'description']]
products["productid"] = products["product_url"].apply(lambda i: crawler.get_ids_from_link(i)[0])
products["shopid"] = products["product_url"].apply(lambda i: crawler.get_ids_from_link(i)[1])
```

```python
k = 1
for i in products["product_url"][0:1]:
    print(f"n_product: {k}")
    print(f"links: {i}")
    shopid, itemid = crawler.get_ids_from_link(i)
    crawler.Crawl(itemid, shopid)
    k += 1
```

```python
df = pd.DataFrame(crawler.data)
df = df[df["comment"] != ""]
```

```python
df.to_excel("file.xlsx", encoding="utf-16")
products.to_excel("./Product info.xlsx", encoding="utf-16")
pd.DataFrame(crawler.GetShopInfo()).to_excel("./Shop info.xlsx", encoding="utf-16")
```
